```python
from bokeh.plotting import figure
from bokeh.io import output_notebook,show
import pandas as pd
from bokeh.models import ColumnDataSource, Div

output_notebook()
```

Dataset location: https://www.kaggle.com/fernandol/countries-of-the-world

```python
countries = pd.read_csv(join(dirname(__file__), 
                             '../datasets/countries_of_the_world.csv'), 
                        decimal=",")
countries.head()
```

```python
countries.describe()
```

```python
regions = list(map(str.strip, list(countries.Region.unique())))
regions.append('ALL')
regions
```

```python
source = ColumnDataSource(data=dict(x=[], 
                                    y=[], 
                                    
                                    color=[], 
                                    country=[], 
                                    gdp=[], 
                                    birthrate=[], 
                                    )
                         )
```

```python
TOOLTIPS = [("Country", "@country"),
            ("GDP per capita", "@gdp"),
            ("Birthrate", "@birthrate")
           ]
```

```python
p = figure(plot_height=300, 
           plot_width=600, 
           
           x_range = (0, 40000), 
           y_range = (0, 55),
           
           title='Countries', 
           
           toolbar_location=None, 
           
           tooltips=TOOLTIPS)
```

```python
p.circle(x="x", 
         y="y", 
         
         source=source, 
         
         size=7,
         line_color=None, 
         )
```

```python
def select_countries():
    
    selected = countries
        
    if (region != 'ALL'):
        selected = selected[selected['Region'] == region]
    
    print('Region = ', region.value)
    
    return selected
```

```python
def update():
    
    df = select_countries()

    #p.xaxis.axis_label = x_axis.value
    #p.yaxis.axis_label = y_axis.value
    p.title.text = 'Countries'
    
    source.data = dict(
        x=df['GDP ($ per capita)'],
        y=df['Birthrate'],
        country=df["Country"],
        gdp=df["GDP ($ per capita)"],
        birthrate=df["Birthrate"]
    )
```

```python
from bokeh.models.widgets import Slider, Select
```

```python
desc = Div(text=open(join(dirname(__file__), "description.html")).read(), width=800)
```

```python
region = Select(title="Region", 
                options=sorted(regions), 
                value='ALL')
```

```python
region.on_change('value', lambda attr, old, new: update())
```

```python
controls = [region]
```

```python
from bokeh.layouts import layout, widgetbox
```

```python
inputs = widgetbox(*controls)
```

```python
plot_layout = layout([
    [desc],
    [inputs, p],
])
```

```python
update()
```

```python
curdoc().add_root(plot_layout)
curdoc().title = "Countries"
```
